```python
#Applying multi-label binarization for Soil_Type (since it can have multiple soil types),

#Label encode Zone, District, Crop, season,

#One-hot encode the same set (Zone, District, Crop, season),

#Save both versions into separate CSV files.
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder

# Load your cleaned dataset
df = pd.read_csv('cleaned_dataset.csv')

# --- 1. Multi-label binarization for 'Soil_Type' ---
# Split soil types by comma and strip extra spaces
df['Soil_Type'] = df['Soil_Type'].apply(lambda x: [soil.strip() for soil in x.split(',')])

mlb = MultiLabelBinarizer()
soil_encoded = pd.DataFrame(mlb.fit_transform(df['Soil_Type']), columns=mlb.classes_)
df_multi = pd.concat([df.drop(columns=['Soil_Type']), soil_encoded], axis=1)

# --- 2. Label encoding ---
df_label = df_multi.copy()
label_cols = ['Zone', 'District', 'Crop', 'season']

label_encoders = {}
for col in label_cols:
    le = LabelEncoder()
    df_label[col] = le.fit_transform(df_label[col])
    label_encoders[col] = le  # optional: store encoders for later decoding

# Save label encoded dataset
df_label.to_csv('final_label_encoded.csv', index=False)

# --- 3. One-hot encoding ---
df_onehot = pd.get_dummies(df_multi, columns=label_cols)

# Save one-hot encoded dataset
df_onehot.to_csv('final_onehot_encoded.csv', index=False)
```
