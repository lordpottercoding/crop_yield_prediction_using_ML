```python
# Check for Extreme Yield Values

def check_extreme_yields(df, column, threshold):
    import matplotlib.pyplot as plt
    import seaborn as sns

    # Find outliers
    extreme_values = df[df[column] > threshold]
    count = extreme_values.shape[0]

    # Print summary
    print(f"⚠️ Found {count} extreme values in '{column}' above threshold {threshold}")
    if count > 0:
        print(extreme_values[[column]].describe())

    # Plotting
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.axvline(x=threshold, color='red', linestyle='--', label=f"Threshold = {threshold}")
    plt.legend()
    plt.show()

    return extreme_values



check_extreme_yields(df, "Kharif_Yield_(Tonne/Hectare)", threshold=15)
check_extreme_yields(df, "Rabi_Yield_(Tonne/Hectare)", threshold=15)
```

```python
# Replace extreme values with the median in the Kharif yield column
def replace_extreme_with_median(df, column, threshold):
    median_value = df[df[column] <= threshold][column].median()
    print(f"Replacing values in '{column}' > {threshold} with median: {median_value}")

    df.loc[df[column] > threshold, column] = median_value
    return df

# Apply to both Kharif and Rabi yield columns
df = replace_extreme_with_median(df, "Kharif_Yield_(Tonne/Hectare)", threshold=15)
df = replace_extreme_with_median(df, "Rabi_Yield_(Tonne/Hectare)", threshold=15)
```

```python
# conforming the replacements 

check_extreme_yields(df, "Kharif_Yield_(Tonne/Hectare)", threshold=15)
check_extreme_yields(df, "Rabi_Yield_(Tonne/Hectare)", threshold=15)
```
